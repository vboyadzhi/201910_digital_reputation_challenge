```python
import numpy as np

sparse_mat = scipy.sparse.coo_matrix((X2.val, (X2.A, X2.id)))
def jaccard_similarities(mat):
    cols_sum = mat.getnnz(axis=0)
    ab = mat.T * mat

    # for rows
    aa = np.repeat(cols_sum, ab.getnnz(axis=0))
    # for columns
    bb = cols_sum[ab.indices]

    similarities = ab.copy()
    similarities.data = similarities.data/(aa + bb - ab.data)

    return similarities

similarities= jaccard_similarities(sparse_mat)
pd.DataFrame(similarities.todense())
```
